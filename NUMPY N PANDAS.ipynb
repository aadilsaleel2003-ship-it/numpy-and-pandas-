```python
#1st question
import numpy as np
import pandas as pd

# Load dataset
df = pd.read_csv("auto-mpg.csv")

mpg_array = df['mpg'].to_numpy()

# Mean, Median, Std
mean_mpg = np.mean(mpg_array)
median_mpg = np.median(mpg_array)
std_mpg = np.std(mpg_array)

print("Mean:", mean_mpg)
print("Median:", median_mpg)
print("Std Dev:", std_mpg)

# Cars with mpg > 25
count = np.sum(mpg_array > 25)
print("Cars with mpg > 25:", count)
```

```python
#2nd question
cars_more_6 = df[df['cylinders'] > 6]['car name'].tolist()
print(cars_more_6)
```

```python
#3rd question
hp = df['horsepower']


hp = pd.to_numeric(hp, errors='coerce')
hp_mean = np.nanmean(hp)

hp_filled = np.where(np.isnan(hp), hp_mean, hp)


hp_new = hp_filled * 1.10

print(hp_new)
```

```python
#4th question
avg_disp = np.mean(df[df['origin'] == 2]['displacement'])
print("Average displacement (Europe):", avg_disp)
```

```python
#5th question
cyl = df['cylinders'].to_numpy()
mpg = df['mpg'].to_numpy()

for c in np.unique(cyl):
    print("Cylinders:", c, "Mean MPG:", np.mean(mpg[cyl == c]))
```

```python
#3
import pandas as pd

# Example: load dataset (replace with your file)
df = pd.read_csv('auto-mpg.csv')

# Convert horsepower to numeric
df['horsepower'] = pd.to_numeric(df['horsepower'], errors='coerce')

# Fill missing values with median
median_hp = df['horsepower'].median()
df['horsepower']=df['horsepower'].fillna(median_hp)

# Check null values
print(df.isnull().sum())
```

```python
#4th
df['power_to_weight_ratio'] = df['horsepower'] / df['weight']
print(df[['power_to_weight_ratio']].head())
```

```python
#5th
grouped = df.groupby('origin')['mpg'].mean()
print(grouped)
```

```python
#6th
summary = df.groupby('model year')[['mpg', 'weight', 'horsepower']].mean()
print(summary)
```

```python
#7th
high_mpg = df[df['mpg'] > 30][['mpg', 'cylinders', 'horsepower', 'weight']]
high_mpg.to_csv("high_mpg_cars.csv", index=False)

print("CSV file saved!")
```
